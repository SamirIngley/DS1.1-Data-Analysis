Activity: Write a function that takes:
1- One of the column (for example Wind Column and its conditions (Weak or Strong) or Temp Column with its conditions (Hot, Mid, Cool)) 
2- The Tennis Player Decision Column
3- Returns probabilities for Tennis Player Decision

```python
def conditional_prob(c1, c2, condition):
    df_new = df[df[c1]==condition][c2]
    total = len(df_new)
    prob = {}
    
    for i in df_new.unique():
        prob[i] = len(df[(df[c1]==condition) & (df[c2] == i)]) / total
    return prob
    
print(conditional_prob('Wind', 'Decision', 'Weak'))
print(conditional_prob('Wind', 'Decision', 'Strong'))
```

```python
print(conditional_prob('Temp', 'Decision', 'Mild'))
```
